```python
Vbase = 2.77128e3
Sbase = 1e6
Ibase = Sbase/Vbase
Zbase = Vbase/Ibase

print(Ibase)

Vdelta = np.zeros((2,3),dtype='complex')
Idelta = np.zeros((2,2),dtype='complex')
Sdelta = np.zeros((2,2),dtype='complex')

print('delta \n ')
dss.run_command('Redirect 02node_delta_mike.dss')
dss.Solution.Solve()

for n in range(len(dss.Circuit.AllBusNames())):
    print(dss.Circuit.AllBusNames()[n])
    dss.Circuit.SetActiveBus(dss.Circuit.AllBusNames()[n])
    print(dss.Bus.kVBase())
    print(dss.Bus.PuVoltage())
    print(dss.Bus.puVmagAngle())

    Vk = np.asarray(dss.Bus.PuVoltage())
    Vk = Vk[0::2] + 1j*Vk[1::2]
    Vk = Vk[0:2]
    Vdelta[:,n] = Vk
    print('')

for n in range(len(dss.Lines.AllNames())):
    print(dss.Lines.AllNames()[n])
    dss.Lines.Name(dss.Lines.AllNames()[n])
    print(dss.CktElement.Currents())
    Ik = np.asarray(dss.CktElement.Currents())
    Ik = (Ik[0::2] + 1j*Ik[1::2])/Ibase
    Ik = -Ik[2:]

    Idelta[:,n] = Ik
    print(Ik)

    print(dss.CktElement.Powers())
    Sk = np.asarray(dss.CktElement.Powers())
    Sk = 1e3*(Sk[0::2] + 1j*Sk[1::2])/Sbase
    Sk = -Sk[2:]

    Sdelta[:,n] = Sk
    print(Sk)

    print('')

    print('Vdelta \n' + str(Vdelta))
    print('Idelta \n' + str(Idelta))
    print('Sdelta \n' + str(Sdelta))
```

```python
Ngen = 20

Npop = 100

Nbest = 10

PQpop = 500*np.random.randn(4,Npop)

Vwye = np.zeros((2,3,Npop),dtype='complex')
Iwye = np.zeros((2,2,Npop),dtype='complex')
Swye = np.zeros((2,2,Npop),dtype='complex')

Jmin = np.zeros(Ngen)
Jmean = np.zeros(Ngen)
Jstd = np.zeros(Ngen)

for kgen in range(0,Ngen):

    Jgen = np.zeros(Npop)


    for kpop in range(0,Npop):


        dss.Loads.Name('load_A02_a_01')
        dss.Loads.kW(PQpop[0,kpop])
        dss.Loads.kvar(PQpop[1,kpop])

        dss.Loads.Name('load_A02_b_01')
        dss.Loads.kW(PQpop[2,kpop])
        dss.Loads.kvar(PQpop[3,kpop])

        dss.Solution.Solve()

        for kbus in range(len(dss.Circuit.AllBusNames())):
            # print(dss.Circuit.AllBusNames()[kbus])
            dss.Circuit.SetActiveBus(dss.Circuit.AllBusNames()[kbus])
            # print(dss.Bus.kVBase())
            # print(dss.Bus.PuVoltage())
            # print(dss.Bus.puVmagAngle())

            Vk = np.asarray(dss.Bus.PuVoltage())
            Vk = Vk[0::2] + 1j*Vk[1::2]
            Vk = Vk[0:2]
            Vwye[:,kbus,kpop] = Vk

        for kline in range(len(dss.Lines.AllNames())):
            # print(dss.Lines.AllNames()[n])
            dss.Lines.Name(dss.Lines.AllNames()[kline])
            # print(dss.CktElement.Currents())
            Ik = np.asarray(dss.CktElement.Currents())
            Ik = (Ik[0::2] + 1j*Ik[1::2])/Ibase
            Ik = -Ik[2:]

            Iwye[:,kline,kpop] = Ik
            # print(Ik)

            # print(dss.CktElement.Powers())
            Sk = np.asarray(dss.CktElement.Powers())
            Sk = 1e3*(Sk[0::2] + 1j*Sk[1::2])/Sbase
            Sk = -Sk[2:]

            Swye[:,kline,kpop] = Sk
            # print(Sk)

            Jgen[kpop] = np.linalg.norm(Vdelta - Vwye[:,:,kpop])**2 + np.linalg.norm(Idelta - Iwye[:,:,kpop])**2

    # Sort SFCs by cost, select the Nbest best
    # print(Jgen)
    idxNbest = Jgen.argsort()[:Nbest]
    # print(idxNbest)

    # Select the nbest best NNs, eliminate rest    
    PQpop = PQpop[:,idxNbest]

    for k1 in range(Nbest):
        for k2 in range(Nbest):
            if k1 != k2:

                # random mutation parameters
                mu = 0
                sigma = 200
                
                # Create new SFC for 10% of cases
                if np.random.rand() <= 0.2:

                    mu = 0
                    sigma = 100
                                        
                    PQpoptemp = np.random.normal(0,100,(4,1))
                    PQpop = np.concatenate((PQpop,PQpoptemp), axis=1)
                
                # Reproduce NNs for 90% of cases
                else:

                    mu = 0
                    sigma = 1

                    # reproduce NNs and add random mutation
                    gamma = np.random.rand(4,1)
                    PQpoptemp = gamma*PQpop[:,k1] + (1 - gamma)*PQpop[:,k2]
                    PQpoptemp += np.random.normal(mu,sigma,(4,1))
                    PQpop = np.concatenate((PQpop,PQpoptemp), axis=1)
    
    # Store cost function values for current generation
    Jmin[kgen] = np.min(Jgen)
    Jmean[kgen] = np.mean(Jgen)
    Jstd[kgen] = np.std(Jgen)
        

    print(kgen)

    #     print(kgen, np.min(JJ), np.mean(JJ), np.std(JJ))
    print('{0:2d}  {1:3.4f}  {2:3.4f}  {3:3.4f}'.format(kgen, np.min(Jgen), np.mean(Jgen), np.std(Jgen)))
#     print('{0:2d}  {1:3.4f}'.format(kgen, np.min(Jgen)))
    print('Best PQ for generation')
    print(PQpop[:,idxNbest[0]])
    print('')

print('PQ')
print(PQpop[:,idxNbest[0]])

print('Vwye')
print(Vwye[:,:,idxNbest[0]])
print('Iwye')
print(Iwye[:,:,idxNbest[0]])
print('Swye')
print(Swye[:,:,idxNbest[0]])

```
